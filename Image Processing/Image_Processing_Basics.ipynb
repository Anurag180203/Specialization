RGB Colour Space

```python
#creating chess in rgb
# 0 -- dark   1 -- light
arr = np.array([[[0,0,0], [255,255,255], [0,0,0]],
                [[255,255,255], [0,0,0], [255,255,255],],
                [[0,0,0], [255,255,255], [0,0,0]]])
plt.imshow(arr)
```

```python
arr = []
for i in range(720):
  temp = []
  for j in range(1280):
    if(i<200):
      temp.append([255,0,0])
    elif(i<500):
      temp.append([0,255,0])
    else:
      temp.append([0,0,255])
  arr.append(temp)
arr = np.array(arr).reshape(720,1280,3)
plt.imshow(arr)
```

```python
cv.imwrite("yo.png",arr)

# while exporting rgb gets converted to bgr
```

```python
img = []

# Red

for i in range(256):             # from Black to Red
    t = []
    for j in range(1536):
        t.append([i,0,0])
    img.append(t)

for i in range(256,0,-1):        # from Red to Black
    t = []
    for j in range(1536):
        t.append([i,0,0])
    img.append(t)

# Green

for i in range(256):             # from Black to Green
    t = []
    for j in range(1536):
        t.append([0,i,0])
    img.append(t)

for i in range(256,0,-1):        # from Green to Black
    t = []
    for j in range(1536):
        t.append([0,i,0])
    img.append(t)

#Blue


for i in range(256):             # from Black to Blue
    t = []
    for j in range(1536):
        t.append([0,0,i])
    img.append(t)

for i in range(256,0,-1):        # from Blue to Black
    t = []
    for j in range(1536):
        t.append([0,0,i])
    img.append(t)



img = np.array(img)

plt.imshow(img)
```

```python
cv.imwrite('Transition.jpg',img)
```
