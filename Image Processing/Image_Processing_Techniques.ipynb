Extracting RGB channel frames

```python
img  = cv.cvtColor(cv.imread('rgb.jpeg'), cv.COLOR_BGR2RGB)
plt.imshow(img)
r = img[:,:,0]
g = img[:,:,1]
b = img[:,:,2]
plt.imshow(r)
```

```python
plt.imshow(g)
```

```python
plt.imshow(b)
```

```python
cv.imwrite('Blue_t.jpg',b)
cv.imwrite('Green_t.jpg',g)
cv.imwrite('Red_t.jpg',r)
```

Showing image in a new window

```python
img = cv.imread('rgb.jpeg') # will work in local systems
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv.imshow("Image",img)
cv.waitKey(0)

img = cv.imread('Transition.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv.imshow("Image",img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img = cv.imread('rgb.jpeg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

r = img[ :  , : , 0 ]  # - Red
g = img[ :  , : , 1 ]  # - Green
b = img[ :  , : , 2 ]  # - Blue


cv.imshow("Frame",r)
cv.waitKey(0)

cv.imshow("Frame",g)
cv.waitKey(0)

cv.imshow("Frame",b)
cv.waitKey(0)
cv.waitKey(0)
cv.destroyAllWindows()
```
